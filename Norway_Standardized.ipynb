```python
MDepth = list(map(float, MDepth))
measuringPoint = list(map(float, measuringPoint))
TVD = map(sub, MDepth, measuringPoint)
```

```python
data_tuples = list(zip(APINumbers, X, Y, (list(TVD))))
```

```python
df_matrix = pd.DataFrame(data_tuples, columns = ['Name', 'X', 'Y', 'True Vertical Depth'])
df_matrix.head
```

```python
df_finalMatrix = pd.DataFrame(df_matrix,  columns = ['Name', 'X', 'Y', 'True Vertical Depth'])
df_finalMatrix.head()
```

```python
listTVD = []
for i in df_finalMatrix['True Vertical Depth']:
    listTVD.append(i)
    
print(listTVD)
```

```python
amin, amax = min(listTVD), max(listTVD)
for i, val in enumerate(listTVD):
    listTVD[i] = (val-amin) / (amax-amin)
```

```python
df_finalMatrix["Normalized TVD"] = pd.Series(listTVD)
```
