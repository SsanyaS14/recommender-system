```python
df1 = pd.read_excel("/Users/samirabr12/Downloads/RecSys 2/Wyoming/CD Files/TeapotDomeWellHeaders02-09-10.xlsx")
#This is what the data initially looks like. As is apparent, we have three different points from which total depth
#is taken : Kelly Bushing, Ground Level, and Drilling Floor elevation. As all datum points are above sea level, we 
# can standardize true vertical depth as (total depth - point of measurement).
print(df1.head())

#Parameters of what we want
#This list contains api numbers of the wells
APINumbers = []
#This list contains the total depths
TotalDepth = []
#This list contains the way the data was measured (e.g. Kelly-Bushing, Ground-Level, Drilling-Floor) 
measuringType = []
#This list contains the value of the point of measurement
measuringPoint = []
northing = []
easting= []
for i in df1["Unnamed: 0"]:
    APINumbers.append(i)
for i in df1["Unnamed: 8"]:
    TotalDepth.append(i)
for i in df1["Unnamed: 10"]:
    measuringType.append(i)
for i in df1["Unnamed: 9"]:
    measuringPoint.append(i)
for i in df1["Unnamed: 5"]:
    easting.append(i)
for i in df1["All Coordinates in Wyoming East Central State Plane, NAD 1927"]:
    northing.append(i)
print(measuringPoint)

northing.remove('Northing')
easting.remove('Easting')
TotalDepth.remove('Total Depth')
measuringPoint.remove('Datum Elevation')
TotalDepth = list(map(float, TotalDepth))
measuringPoint = list(map(float, measuringPoint))
TVD = map(sub, TotalDepth, measuringPoint)
APINumbers.remove('API Number')
data_tuples = list(zip(APINumbers, northing, easting, (list(TVD))))

df1_matrix = pd.DataFrame(data_tuples, columns = ['Name', 'Northing', 'Easting', 'True Vertical Depth'])
df1_matrix.head
df1_finalMatrix = pd.DataFrame(df1_matrix,  columns = ['Name', 'Northing', 'Easting', 'True Vertical Depth'])
df1_finalMatrix.head()
listTVD = []
for i in df_finalMatrix['True Vertical Depth']:
    listTVD.append(i) 
print(listTVD)
amin, amax = min(listTVD), max(listTVD)
for i, val in enumerate(listTVD):
    listTVD[i] = (val-amin) / (amax-amin)
df1_finalMatrix["Normalized TVD"] = pd.Series(listTVD)
df1_finalMatrix.head()
```
